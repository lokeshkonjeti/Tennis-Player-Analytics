```python
import cv2
from ultralytics import YOLO

model = YOLO("weights/best.pt") #Replace the path for the weights file

video_path = "cv_project/input_video.mp4"  # Replace with your video path
cap = cv2.VideoCapture(video_path)

fps = cap.get(cv2.CAP_PROP_FPS)
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

fourcc = cv2.VideoWriter_fourcc(*'mp4v')
output_path = "output_videos/output_video.mp4". #Specify the output path
out = cv2.VideoWriter(output_path, fourcc, fps, (width, height))

player_ids = {}
next_player_id = 1

while cap.isOpened():
    ret, frame = cap.read()
    if ret:
        results = model(frame)
        annotated_frame = results[0].plot()
        for box in results[0].boxes:
            cls = int(box.cls)
            confidence = box.conf
            xyxy = box.xyxy[0].tolist()
            x1, y1, x2, y2 = map(int, xyxy)
            object_id = None
            for existing_id, (existing_x1, existing_y1, existing_x2, existing_y2) in player_ids.items():
                x_left = max(x1, existing_x1)
                y_top = max(y1, existing_y1)
                x_right = min(x2, existing_x2)
                y_bottom = min(y2, existing_y2)
                intersection_area = max(0, x_right - x_left) * max(0, y_bottom - y_top)
                box_area = (x2-x1) * (y2-y1)
                existing_box_area = (existing_x2 - existing_x1) * (existing_y2-existing_y1)
                union_area = box_area + existing_box_area - intersection_area
                iou = intersection_area / union_area if union_area > 0 else 0
                if iou > 0.5:
                    object_id = existing_id
                    player_ids[object_id] = (x1, y1, x2, y2)
                    break
            if object_id is None:
                object_id = next_player_id
                player_ids[object_id] = (x1, y1, x2, y2)
                next_player_id += 1
            cv2.putText(annotated_frame, f"Player {object_id}", (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 255, 0), 2)

        out.write(annotated_frame)
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
    else:
        break

cap.release()
out.release()
cv2.destroyAllWindows()

print(f"Tracking complete. Output video saved to: {output_path}")
```
